# **코랩에서 데이터 확인하기**

```python
import gdown
gdown.download('https://bit.ly/3eecMKZ',
               '남산도서관 장서 대출목록 (2021년 04월).csv', quiet=False)
# 웹에서 다운받은 데이터
```

```python
import chardet
with open('남산도서관 장서 대출목록 (2021년 04월).csv', mode='rb') as f:
    d = f.readline()

print(chardet.detect(d))

# mode='rb' 읽기 모드를 바이너리 모드로 지정해서 읽을 때 오류나지 않게 한 것.
```

```python
with open('남산도서관 장서 대출목록 (2021년 04월).csv', encoding='euc-kr') as f:
    print(f.readline())
    print(f.readline())
    print(f.readline())
# print(f.readline()) 입력하는 개수만큼 위에서부터 줄 보여줌
```

# **데이터프레임 다루기: 판다스**

```python
import pandas as pd
df = pd.read_csv('남산도서관 장서 대출목록 (2021년 04월).csv', encoding='euc-kr')
# 판다스로 csv 파일 읽겠다.
```

```python
df = pd.read_csv('남산도서관 장서 대출목록 (2021년 04월).csv', encoding='euc-kr',
                 low_memory=False)
# 데이터 타입을 헷갈리지 않게 한번에 불러와줄 수 있도록 한 것. 데이터를 조각으로 읽어오지 않게 한 것.
# dtype={'ISBN': str, '세트 ISBN': str, '주제분류번호': str}) 아니면 이런 식으로 dtype 매개변수로 데이터 타입 지정해줄 수도 있음.
```

```python
df.head()
# 데이터 프레임의 윗부분 확인 가
```

```python
df.to_csv('ns_202104.csv')
# 파일 창의 디렉터리에 'ns_202104.csv' 이름의 파일이 생성됨
```
